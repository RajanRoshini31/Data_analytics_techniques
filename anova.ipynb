```python
from scipy.stats import f_oneway

print("One-Way ANOVA Test")

# Enter data for 3 groups
g1 = list(map(float, input("Enter Group 1 values: ").split()))
g2 = list(map(float, input("Enter Group 2 values: ").split()))
g3 = list(map(float, input("Enter Group 3 values: ").split()))

# Perform ANOVA
f_stat, p= f_oneway(g1, g2, g3)

# Input significance level
alpha = float(input("Enter significance level (e.g., 0.05): "))

# Output
print("\nF-statistic:", round(f_stat, 4))

if p < alpha:
    print("Result: Reject Null Hypothesis  At least one group is different.")
else:
    print("Result: Fail to Reject Null Hypothesis  All groups are similar.")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

print("Two-Way ANOVA Test")

# Enter your data directly
# Example: Exam scores by two factors Ã¢â¬â Gender and Method
data = {
    'Score': [80, 85, 78, 82, 90, 88],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Method': ['A', 'A', 'B', 'B', 'A', 'B']
}

# Create DataFrame
df = pd.DataFrame(data)

# Show data
print("\n--- Input Data ---")
print(df)

# Build the model with interaction
model = ols('Score ~ C(Gender) + C(Method) + C(Gender):C(Method)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Significance level input
alpha = float(input("\nEnter significance level (e.g., 0.05): "))

# Show ANOVA table
print("\n--- Two-Way ANOVA Results ---")
print(anova_table)

# Interpret results
for factor in anova_table.index:
    p = anova_table.loc[factor, 'PR(>F)']
    if p < alpha:
        print(f"Result: Reject Null Hypothesis for '{factor}'  it has a significant effect.")
    else:
        print(f"Result: Fail to Reject Null Hypothesis for '{factor}' no significant effect.")
```
