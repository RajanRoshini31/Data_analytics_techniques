```python
import pandas as pd
df=pd.read_csv('/content/CSM_A&B.csv')
df
```

```python

columns_to_drop = []

for column_name in df.columns:
    null_count = df[column_name].isnull().sum()
    total_rows = len(df)
    null_percentage = (null_count / total_rows) * 100
    if null_percentage > 50:
        columns_to_drop.append(column_name)

df.drop(columns=columns_to_drop, inplace=True)
print("Columns dropped:", columns_to_drop)
df
```

```python
import numpy as np
from scipy import stats

def t_test(sample1, sample2=None, paired=False, alternative='two-sided', alpha=0.05):

    if sample2 is None: # One sample t-test
        t_statistic, p_value = stats.ttest_1samp(sample1, popmean=np.mean(sample1) , alternative=alternative)
    elif paired: # Paired t-test
         t_statistic, p_value = stats.ttest_rel(sample1, sample2, alternative=alternative)
    else:  # Two sample t-test
        t_statistic, p_value = stats.ttest_ind(sample1, sample2, alternative=alternative)

    return t_statistic, p_value

# Example Usage:
# One sample t-test:
sample1 = np.array([60, 72, 65, 75, 80, 78, 70, 75, 68])
t_stat, p_val = t_test(sample1, alternative = 'greater')
print("One-Sample T-test:")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_val:.3f}")

alpha = 0.05
if p_val < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
print("\n")
# Two sample t-test:
sample_a = np.array([60, 72, 65, 75, 80, 78, 70, 75, 68])
sample_b = np.array([55, 68, 60, 70, 74, 72, 65, 70, 62])

t_stat, p_val = t_test(sample_a, sample_b, alternative = 'two-sided')
print("Two-Sample T-test:")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_val:.3f}")

if p_val < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
print("\n")
# Paired t-test:
before = np.array([10, 12, 15, 13, 16, 18, 17, 15, 14])
after = np.array([12, 14, 16, 15, 18, 20, 19, 17, 15])

t_stat, p_val = t_test(before, after, paired=True, alternative = 'less')

print("Paired T-test:")
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_val:.3f}")


if p_val < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
import numpy as np
from scipy import stats

def z_test(sample_mean, pop_mean, pop_std, sample_size, alternative='two-sided'):
    z_statistic = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    return z_statistic, p_value

# Example usage:
sample_mean = 110  # Sample mean
pop_mean = 100    # Population mean
pop_std = 15      # Population standard deviation
sample_size = 30   # Sample size

# Perform the z-test
z_statistic, p_value = z_test(sample_mean, pop_mean, pop_std, sample_size)

# Print the results
print("Z-statistic:", z_statistic)
print("P-value:", p_value)

# Interpret the results (example)
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```
